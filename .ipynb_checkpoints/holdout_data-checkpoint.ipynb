```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import f1_score
import pickle
%matplotlib inline

pd.options.display.max_columns = 100
```

```python
holdout_data = pd.read_csv('holdout_data.csv')
holdout_data.drop(columns=['Unnamed: 0'])
```

```python
holdout_data.rename(columns = {"X1": "credit_amount", 
                     "X2":"gender", 
                     "X3":"education", 
                     "X4":"marital_status", 
                     "X5":"age", 
                     "X6":"repay_status_sept",
                     "X7":"repay_status_aug",
                     "X8":"repay_status_jul",
                     "X9":"repay_status_jun",
                     "X10":"repay_status_may",
                     "X11":"repay_status_apr",
                    "X12":"statement_amt_sept",
                    "X13":"statement_amt_aug",
                    "X14":"statement_amt_jul",
                    "X15":"statement_amt_jun",
                    "X16":"statement_amt_may",
                    "X17":"statement_amt_apr",
                    "X18":"amt_paid_sept",
                    "X19":"amt_paid_aug",
                    "X20":"amt_paid_jul",
                    "X21":"amt_paid_jun",
                    "X22":"amt_paid_may",
                    "X23":"amt_paid_april",
                    "Y":"defaulted"}, inplace = True) 
holdout_data.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
conditions = [
    (holdout_data['marital_status'] == 1),
    (holdout_data['marital_status'] != 1)
]

values = [1, 0]
holdout_data['is_married'] = np.select(conditions, values)
```

# Feature Eng

```python
holdout_data['outstanding'] = holdout_data['statement_amt_sept'] + holdout_data['statement_amt_aug'] + holdout_data['statement_amt_jul'] + holdout_data['statement_amt_jun'] + holdout_data['statement_amt_may'] + holdout_data['statement_amt_apr']
holdout_data['paid'] = holdout_data['amt_paid_sept'] + holdout_data['amt_paid_aug'] + holdout_data['amt_paid_jul'] + holdout_data['amt_paid_jun'] + holdout_data['amt_paid_may'] + holdout_data['amt_paid_april']

holdout_data['balance'] = holdout_data['outstanding'] - holdout_data['paid']
holdout_data['utilization'] = (holdout_data['balance']/holdout_data['credit_amount'])*100
```

```python
holdout_data
```

# Dummies

```python
gender_dummies = pd.get_dummies(holdout_data['gender'], prefix='gender', drop_first=True)
education_dummies = pd.get_dummies(holdout_data['education'], prefix='education', drop_first=True)
marital_status_dummies = pd.get_dummies(holdout_data['is_married'], prefix='married', drop_first=True)
sep_repay = pd.get_dummies(holdout_data['repay_status_sept'], prefix='repay_sep', drop_first=True)
aug_repay = pd.get_dummies(holdout_data['repay_status_aug'], prefix='repay_aug', drop_first=True)
jul_repay = pd.get_dummies(holdout_data['repay_status_jul'], prefix='repay_jul', drop_first=True)
jun_repay = pd.get_dummies(holdout_data['repay_status_jun'], prefix='repay_jun', drop_first=True)
may_repay = pd.get_dummies(holdout_data['repay_status_may'], prefix='repay_may', drop_first=True)
aug_repay = pd.get_dummies(holdout_data['repay_status_apr'], prefix='repay_apr', drop_first=True)

holdout_data = holdout_data.drop(['gender', 
                                    'education', 
                                    'marital_status',
                                    'is_married',
                                    'repay_status_sept',
                                    'repay_status_aug',
                                    'repay_status_jul',
                                    'repay_status_jun',
                                    'repay_status_may',
                                    'repay_status_apr'

                                    ], axis=1)


holdout_data = pd.concat([holdout_data, 
                           gender_dummies, 
                           education_dummies, 
                           marital_status_dummies,
                           sep_repay,
                           aug_repay,
                           jul_repay,
                           jun_repay,
                           may_repay,
                           aug_repay], axis=1)
```
